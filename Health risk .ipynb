```python
import numpy as np
import pandas as pd
```

```python
df_data=pd.read_csv('Health Risk Data Set.csv',encoding='latin-1')
df_data.info()
df_data.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
vector=CountVectorizer()
```

```python
x=df_data.drop(columns=['RiskLevel'])
y=df_data['RiskLevel']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=3)
print(df_data.columns)
X = df_data[['ï»¿Age','SystolicBP','DiastolicBP','BS','BodyTemp','HeartRate']]
Y= df_data['RiskLevel'] 
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

```

```python
from sklearn .preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(random_state=42)
model.fit(X_train,Y_train)
```

```python
y_predict=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,y_predict)
print("Accuracy:",accuracy)
```

```python
import pickle
with open('Health Risk Data Set.pkl','wb') as file:
  pickle.dump(model,file)
with open('scaler.pkl','wb') as sc_file:
  pickle.dump(scaler,sc_file)
```
